```python
columns = ['TV', 'Radio', 'Newspaper', 'Sales']
pre_a_df = a_df[columns]
pre_a_df
```

```python
import matplotlib.pyplot as plt

# 'subplots'를 사용하여 여러 subplot 생성
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 여기서 'axes'는 각 subplot의 축 객체 배열입니다.

# 첫 번째 subplot에 대한 산점도 그리기
axes[0].scatter(pre_a_df['TV'], pre_a_df['Sales'])

# 두 번째 subplot에 대한 산점도 그리기
axes[1].scatter(pre_a_df['Radio'], pre_a_df['Sales'])

# 세 번째 subplot에 대한 산점도 그리기
axes[2].scatter(pre_a_df['Newspaper'], pre_a_df['Sales'])

# 그래프 표시
plt.show()
```

```python
# TV
```

```python
pre_tv_df = pre_a_df[['TV', 'Sales']]
pre_tv_df
```

```python
pre_tv_df.hist()
```

```python
from torch.nn import Module, Linear
class LinearRegressionModel(Module):
    def __init__(self):
        super().__init__()
        self.linear = Linear(1,1)
    def forward(self, x):
        return self.linear(x)
```

```python
from sklearn.model_selection import train_test_split
import torch
from torch.nn.functional import mse_loss
from torch.optim import SGD

torch.manual_seed(124)

features, targets = pre_tv_df.iloc[:, :-1], pre_tv_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

X_train = torch.FloatTensor(X_train.values)
y_train = torch.FloatTensor(y_train.values).view(-1, 1)

X_test = torch.FloatTensor(X_test.values)
y_test = torch.FloatTensor(y_test.values).view(-1, 1)

l_r = LinearRegressionModel()
optimizer = SGD(l_r.parameters(), lr=1e-6)

epochs = 100000

for epoch in range(1, epochs + 1):
    H = l_r(X_train)
    loss = mse_loss(H, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        W = list(l_r.parameters())[0].item()
        b = list(l_r.parameters())[1].item()
        print('{:4d}/{}: W: {:.4f}, b: {:.4f}, loss: {:.4f}'\
              .format(epoch, epochs, W, b, loss.item()))
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

plt.scatter(X_train, y_train)
plt.plot(X_train, 0.0803 * X_train + 0.5781, color="red")
plt.grid(visible=True, linestyle='--')
# plt.legend()
plt.show()
```

```python
import numpy as np

H = 0.0803 * X_test + 0.5781  # 여기서 X_train 대신 X_test 사용

# RMSE 계산
print(np.sqrt(mse_loss(H, y_test).item()))
```
