```python
pre_df = df.copy()
pre_df
```

```python
pre_df = pre_df.drop(labels=['ID'], axis=1)
pre_df
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['EDUCATION', 'default ']
label_encoders = {}

for column in columns:
    encoder = LabelEncoder()
    result = encoder.fit_transform(pre_df[column])
    label_encoders[column] = encoder.classes_
    pre_df[column] = result

pre_df
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# sparse를 False로 설정하면 넘파이 배열, True는 희소 행렬
one_hot_encoder = OneHotEncoder(sparse_output=False)
result = one_hot_encoder.fit_transform(pre_df[['MARRIAGE']])

pre_df = pd.concat([pre_df, 
                      pd.DataFrame(result, columns=one_hot_encoder.categories_).astype(np.int8)], 
                     axis=1)
pre_df
```
