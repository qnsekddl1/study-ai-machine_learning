```python
import pandas as pd

df = pd.read_csv('./datasets/syn_crab_data (2).csv')

df
```

```python
df.info()
```

```python
df.Sex.value_counts()
```

```python
df = df[df['Sex'] != 0.0]
df
```

```python
df = df.drop(labels=['id'], axis=1)
df
```

```python
pre_df = df.copy()
pre_df = pre_df[pre_df['Sex'] != 'I']
```

```python
pre_df
```

```python
pre_df.Sex.value_counts()
```

```python
pre_df = pre_df[(pre_df['Sex'] == 'M') | (pre_df['Sex'] == 'F')]
pre_df
```

```python
pre_df = pre_df.reset_index(drop=True)
pre_df
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['Sex']
label_encoders = {}

for column in columns:
    encoder = LabelEncoder()
    result = encoder.fit_transform(pre_df[column])
    label_encoders[column] = encoder.classes_
    pre_df[column] = result

pre_df
```

```python
from statsmodels.api import OLS

features, targets = pre_df.iloc[:, :-1], pre_df.iloc[:, -1]

model = OLS(targets, features)
print(model.fit().summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
get_vif(features)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, MSLE, RMSLE, R2))
    # print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, R2))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features, targets = pre_df.iloc[:, :-1], pre_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
prediction = l_r.predict(X_test)
get_evaluation(y_test, prediction)
```
