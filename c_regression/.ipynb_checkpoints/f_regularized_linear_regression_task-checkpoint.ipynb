#### Regularized Linear Regression  
- 다이아몬드의 특성을 회귀분석하여, 예상 가격을 예측한다.

```python
import pandas as pd

df = pd.read_csv('./datasets/diamond.csv')
df
```

```python
pre_df = df.drop(labels=['Unnamed: 0'], axis=1)
pre_df
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['cut', 'color', 'clarity']
label_encoders = {}

for column in columns:
    encoder = LabelEncoder()
    result = encoder.fit_transform(pre_df[column])
    label_encoders[column] = encoder.classes_
    pre_df[column] = result

label_encoders
```

```python
pre_df
```

```python
pre_df['target'] = pre_df.price
pre_df = pre_df.drop(labels=['price'], axis=1)
pre_df
```

```python
import statsmodels.api as sm

features, targets = pre_df.iloc[:, :-1], pre_df.target

print(sm.OLS(targets, features).fit().summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def feature_engineering_VIF(features):
    vif = pd.DataFrame()
    vif['score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
display(feature_engineering_VIF(features))
```
