```python
import pandas as pd

df = pd.read_csv('./datasets/ex_contests2.csv')
df
```

```python
pre_df = df.copy()
pre_df
```

```python
columns = ['Title', 'Content']
```

```python
features = pre_df.loc[:, columns]
```

```python
def concatenate(features):
    return features.Title + ' ' + features.Content
result_df = concatenate(features)
result_df 
```

```python
pre_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_v = CountVectorizer()
count_metrix = count_v.fit_transform(result_df)
print(count_metrix)

c_s = cosine_similarity(count_metrix)
print(c_s)
print(c_s.shape)
```
