```python
import pandas as pd
import numpy as np

df = pd.read_csv('./datasets/ex_contests2.csv')

df
```

```python
df.isna().sum()
```

```python
pre_df = df.copy()
pre_df
```

```python
columns = ['Title', 'Content']
```

```python
features = pre_df.loc[:, columns]
```

```python
features
```

#### 1. 제목 + 전체 내용

```python
def concatenate(features):
    return features.Title + ' ' + features.Content

result_df = concatenate(features)
result_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_v = CountVectorizer()
count_metrix = count_v.fit_transform(result_df)
print(count_metrix)

c_s = cosine_similarity(count_metrix)
print(c_s)
print(c_s.shape)
```

```python
import pandas as pd

# 특정 제목으로 인덱스 찾기
def get_index_from_title(title):
    matching_titles = pre_df[pre_df['Title'].str.contains(title, na=False)]
    if not matching_titles.empty:
        return matching_titles.index[0]
    else:
        raise ValueError(f"Title containing '{title}' not found in the DataFrame")

# 특정 인덱스로 제목 찾기
def get_title_from_index(index):
    if index in df.index:
        return df.loc[index, 'Title']
    else:
        raise IndexError(f"Index '{index}' is out of bounds")

# 설정된 제목을 기반으로 추천 제목 출력
exhibition_title = "2024"
try:
    exhibition_index = get_index_from_title(exhibition_title)
    recommended_exhibition = sorted(list(enumerate(c_s[exhibition_index])), key=lambda x: x[1], reverse=True)
    
    # 전체 글자 출력 설정
    pd.set_option('display.max_colwidth', None)

    # 추천 게시글 출력
    for exhibition in recommended_exhibition:
        print(f"({exhibition[0]}, {exhibition[1]})")
    print("\nRecommended Titles:")
    for exhibition in recommended_exhibition[1:6]:
        print(get_title_from_index(exhibition[0]))
except (ValueError, IndexError) as e:
    print(e)
```

```python
# Content 열에서 첫 두 단어만 추출하는 함수 정의
def first_two_words(content):
    return ' '.join(content.split()[:2])

# features DataFrame에 apply를 사용하여 첫 두 단어만 추출
features['ShortContent'] = features['Content'].apply(first_two_words)

# Title과 ShortContent를 결합하는 함수 정의
def concatenate(features):
    return features.Title + ' ' + features.ShortContent

# 열 결합
result_df = concatenate(features)
result_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_v = CountVectorizer()
count_metrix = count_v.fit_transform(result_df)
print(count_metrix)

c_s = cosine_similarity(count_metrix)
print(c_s)
print(c_s.shape)
```
