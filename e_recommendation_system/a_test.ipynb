```python
import numpy as np
import pandas as pd

df = pd.read_csv('./datasets/wiki_movie_plots_deduped.csv')

df
```

```python
df.isna().sum()
```

```python
pre_df = df.copy()
pre_df
```

```python
df.info()
```

```python
# Genre Title Director Plot Cast
columns = ['Genre', 'Title', 'Director', 'Plot', 'Cast']
```

```python
pre_df = pre_df[columns].fillna('')
pre_df.isna().sum()
```

```python
features = pre_df.loc[:, columns]
```

```python
def concatenate(features):
    return features.Title + ' ' + features.Genre + \
    features.Director + ' ' + features.Plot + ' ' + features.Cast
result_df = concatenate(features)
result_df 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_v = CountVectorizer()
count_metrix = count_v.fit_transform(result_df)
print(count_metrix)

c_s = cosine_similarity(count_metrix)
print(c_s)
print(c_s.shape)
```

```python
def get_index_from_title(title) : 
    return df[df.Title == title].index[0]


def get_title_from_index(index) :
    return df[df.index == index]['Title'].values[0]

# get_title_from_index(2)
movie_title = "The Water Diviner"
movie_index = get_index_from_title(movie_title)
recommended_movie = sorted(list(enumerate(c_s[movie_index])), key=lambda x:x[1], reverse=True)
recommended_movie

for movie in recommended_movie[1:6] :
    print(get_title_from_index(movie[0]))
```
