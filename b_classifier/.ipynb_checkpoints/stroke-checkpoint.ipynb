```python
import pandas as pd
import numpy as np

hd_df = pd.read_csv('./datasets/healthcare_dataset_stroke_data.csv')
hd_df
```

```python
mean_bmi = hd_df['bmi'].mean()
```

```python
hd_df['bmi'].fillna(mean_bmi, inplace=True)
```

```python
hd_df.isna().sum()
```

```python
hd_df.drop(columns=['id'], inplace=True)
```

```python
hd_df
```

```python
hd_enc_df = hd_df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

hd_encoder = LabelEncoder()

# 파이썬 리스트로 전달해야 한다.
targets = hd_encoder.fit_transform(hd_df['stroke'].tolist())
hd_enc_df['stroke'] = targets

gender_encoder = LabelEncoder()
genders = gender_encoder.fit_transform(hd_df['gender'].tolist())
hd_enc_df['gender'] = genders

married_encoder = LabelEncoder()
marrrieds = married_encoder.fit_transform(hd_df['ever_married'].tolist())
hd_enc_df['ever_married'] = marrrieds

work_type_encoder = LabelEncoder()
work_types = work_type_encoder.fit_transform(hd_df['work_type'].tolist())
hd_enc_df['work_type'] = work_types

residence_type_encoder = LabelEncoder()
residence_types = residence_type_encoder.fit_transform(hd_df['Residence_type'].tolist())
hd_enc_df['Residence_type'] = residence_types

smoking_status_encoder = LabelEncoder()
smoking_statuses = smoking_status_encoder.fit_transform(hd_df['smoking_status'].tolist())
hd_enc_df['smoking_status'] = smoking_statuses
```

```python
hd_enc_df
```

```python
hd_enc_df.info()
```

```python
hd_enc_df['age'] = hd_enc_df['age'].astype('int64')
hd_enc_df['avg_glucose_level'] = hd_enc_df['avg_glucose_level'].astype('int64')
hd_enc_df['bmi'] = hd_enc_df['bmi'].astype('int64')
```

```python
hd_enc_df.info()
```

```python
import matplotlib.pyplot as plt

plt.hist(hd_enc_df['age'], bins=100, color="green")
plt.show()

plt.hist(hd_enc_df['avg_glucose_level'], bins=100, color="blue")
plt.hist(hd_enc_df['bmi'], bins=100, color="skyblue")
plt.show()
```

```python
# 이상치 확인

import numpy as np

zero_features = ['age', 'ever_married', 'work_type', 'Residence_type', 'avg_glucose_level', 'smoking_status', 'bmi']

total_count = hd_enc_df.shape[0]

for feature in zero_features:
    zero_count = hd_df[hd_df[feature] == 0][feature].count()
    print(f'{feature}: {zero_count}건, {np.round(zero_count / total_count * 100, 4)}%')
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.tree import DecisionTreeClassifier

features, targets = hd_enc_df.iloc[:, :-1], hd_enc_df.iloc[:, -1]

# stratify: 데이터 분리 시 타겟 데이터의 분포 비율에 맞춰서 나눠준다.
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, stratify=targets, random_state=124)

dtc = DecisionTreeClassifier()
parameters = {'max_depth': range(3, 10), 
              'min_samples_split': range(12, 20)
                  }

g_dtc = GridSearchCV(dtc, param_grid=parameters, cv=5, refit=True, return_train_score=True)

# 교차 검증 진행
g_dtc.fit(X_train, y_train)

pd.DataFrame(g_dtc.cv_results_)[['params', 'mean_test_score', 'rank_test_score', 
           'split0_test_score', 'split1_test_score', 'split2_test_score']]
```

```python
from sklearn.metrics import accuracy_score

dtc = g_dtc.best_estimator_
prediction = dtc.predict(X_test)

print(f'정확도: {np.round(accuracy_score(y_test, prediction) * 100, 4)}%')
print(g_dtc.best_params_, g_dtc.best_score_, sep="\n")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# 타겟 데이터와 예측 객체를 전달받는다.
def get_evaluation(y_test, prediction, classifier=None, X_test=None):
#     오차 행렬
    confusion = confusion_matrix(y_test, prediction)
#     정확도
    accuracy = accuracy_score(y_test , prediction)
#     정밀도
    precision = precision_score(y_test , prediction)
#     재현율
    recall = recall_score(y_test , prediction)
#     F1 score
    f1 = f1_score(y_test, prediction)
#     ROC-AUC
    roc_auc = roc_auc_score(y_test, prediction)
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
get_evaluation(y_test, prediction, dtc, X_test)
```

```python
proba = dtc.predict_proba(X_test)
prediction = dtc.predict(X_test)

for proba, predict in zip(proba.tolist(), prediction.tolist()):
    print(proba, predict)
```
